```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from imblearn.over_sampling import SMOTE
```

```python
def readdata(file):
    df = pd.read_csv(file)
    targets = df['Threat']
    attributes = df.drop('Threat', axis=1)
    return attributes, targets
attributes, targets = readdata('threats.csv')
attributes.head()
```

```python
targets
```

```python
def oversampling(attributes, targets):
    smote = SMOTE()
    attributes, targets = smote.fit_resample(attributes, targets)
    return attributes, targets
attributes, targets = oversampling(attributes, targets)
```

```python
def splitter(attributes, targets):
    train_attributes, test_attributes, train_targets, test_targets \
        = train_test_split(attributes, targets)
    return train_attributes, test_attributes, train_targets, test_targets
train_attributes, test_attributes, train_targets, test_targets = splitter(attributes, targets)
```

```python
def trainmodel(train_attributes, train_targets):
    clf = DecisionTreeClassifier()
    clf.fit(train_attributes, train_targets)
    return clf
clf = trainmodel(train_attributes, train_targets)
```

```python
def testmodel(test_attributes, test_targets, clf):
    predictions = clf.predict(test_attributes)
    report = classification_report(test_targets, predictions)
    return report
rep = testmodel(test_attributes, test_targets, clf)
print(rep)
```
